```python
from PIL import Image as PImage
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Test-1.jpg"
img = PImage.open(path)
```

```python
plt.imshow(img)
plt.show()
```

```python
(layerR, layerG, layerB) = img.split()
layerRn = np.array(layerR)
layerGn = np.array(layerG)
layerBn = np.array(layerB)
```

```python
rowLayer = layerR.size[0]
colLayer = layerR.size[1]
```

```python
layerRn = layerRn.reshape(rowLayer*colLayer, 1)
layerGn = layerGn.reshape(rowLayer*colLayer, 1)
layerBn = layerBn.reshape(rowLayer*colLayer, 1)
dataLayer = np.column_stack((layerRn, layerGn, layerBn))
```

```python
Kmeans = KMeans(n_clusters = 9, random_state = 9)
Kmeans.fit(dataLayer)
predicted = Kmeans.predict(dataLayer)
```

```python
imgNew = np.array(predicted)
imgNew = predicted.reshape(colLayer, rowLayer)
```

```python
imgNew = (imgNew*255).astype(np.uint8)
imgNew = PImage.fromarray(imgNew)
```

```python
plt.imshow(imgNew)
plt.show()
```
